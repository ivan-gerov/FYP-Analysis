```python
alphas = pd.read_csv(os.path.join(data_path, 'alpha_all_ascii.csv'),
                    sep= '\t',
                    names = ['silence', 'white', 'lofi']
                    )
```

```python
# alphas.drop(columns= ['Unnamed: 3'], inplace= True)
```

```python
alphas.reset_index(inplace= True)
```

```python
alphas.rename(columns = {
    'index': 'silence',
    'silence': 'white',
    'white': 'lofi',
    'lofi': 'remove'
}, inplace= True)
```

```python
alphas.drop(columns = ['remove'], inplace= True)
```

```python
alpha_contra = alphas[:11].reset_index(drop=True)
alpha_ipsi = alphas[11:].reset_index(drop=True)

print(alpha_contra.shape, alpha_ipsi.shape)
```

```python
alpha_contra.rename(columns = {
    'silence': 'contra_silence',
    'white': 'contra_white',
    'lofi': 'contra_lofi'
}, inplace= True)

alpha_ipsi.rename(columns = {
    'silence': 'ipsi_silence',
    'white': 'ipsi_white',
    'lofi': 'ipsi_lofi'
}, inplace= True)
```

```python
alpha_reshaped = pd.concat([alpha_contra, alpha_ipsi], axis=1)
```

```python
alpha_reshaped.to_csv('./alpha_reshaped.csv',
                 index= False,
                 encoding= 'utf8')
```
